```python
class DynamicArray:
    """A dynamic array class akin to a simplified Python list."""
    def getsize(self):
        import sys #Bunun bir fonksiyonu var class oldugunda mutlaka metodun ilk paydası self olmak zorunda.
        try:
            return sys.getsizeof(self._A)
          
        except:
            return 0
    def ToString(self):
        try:        #Hataya karşı try-except ekledik.
            for i in self._A: #A dediği dizinin kendisi.
                print(i," ")
        except:
            pass
    
    def getLength(self):
        return len(self._A)
    def __init__(self):
        """Create an empty array."""
        self._n = 0        #n yi 0 olarak belirledi # count actual elements
        self._capacity = 1  #kapasiteyi 1 olarak belirledi.  # default array capacity
        self._A = self._make_array(self._capacity)     # low-level array
    def _make_array(self, c): #c dediği bana gönderilen değer ve c yi 1 gönderdi.  # nonpublic utitity
        """Return new array with capacity c."""  
        return (c * ctypes.py_object)()    #1 birimlik yer ayırıp bunun değerini bana gönderdi 1*ctypes.py_object boyutu kadar bana bir yer sağlar bellekte.
    def append(self, obj):
        """Add object to end of the array."""
        if self._n == self._capacity:  #Kapasite ile n aynı yer mi? # not enough room
            self._resize(2 * self._capacity) #Bu kadar resize yapıyor.  #Aynı yer ise dolmuş demektir. Liste için bir resize yap.# so double capacity
        self._A[self._n] = obj #Bulundugun n değerine o an ki objeyi ekle.
        self._n += 1 #n yi bir arttır. n neydi kaç eleman var onu gösteriyordu. Kapasitede toplam boyutu söyler.
    def _resize(self, c):    #c diye bir değ alıyor. # nonpublic utitity
        """Resize internal array to capacity c."""
        B = self._make_array(c)     #c*2=.. lik bir yer ayır.     # new (bigger) array
        print(" şu an amortized cost işlemi ... ")
        for k in range(self._n):                # for each existing value
            B[k] = self._A[k]
            print(" şu an move işlemi ... ")
        self._A = B                     # use the bigger array
        self._capacity = c #Kapasitem 2*c kadar oldu.

    def __len__(self): # Kapasiteyi
        """Return number of elements stored in the array."""
        return self._n
    def len_n(self):
        """Return number of elements stored in the array."""
        return self._n
    
    def __getitem__(self, k): #K indisi alıyor ve oradaki veriyi bana söylüyor.
        """Return element at index k."""
        if not 0 <= k < self._n:
            raise IndexError('invalid index')
        return self._A[k]                              # retrieve from array
    def insert(self, k, value):
        """Insert value at index k, shifting subsequent values rightward."""
        # (for simplicity, we assume 0 <= k <= n in this verion)
        if self._n == self._capacity:                  # not enough room
            self._resize(2 * self._capacity)             # so double capacity
        for j in range(self._n, k, -1):                # shift rightmost first
            self._A[j] = self._A[j-1]
        self._A[k] = value                             # store newest element
        self._n += 1

    def remove(self, value):
        """Remove first occurrence of value (or raise ValueError)."""
        # note: we do not consider shrinking the dynamic array in this version
        for k in range(self._n):
            if self._A[k] == value:              # found a match!
                for j in range(k, self._n - 1):    # shift others to fill gap
                    self._A[j] = self._A[j+1]
                self._A[self._n - 1] = None        # help garbage collection
                self._n -= 1                       # we have one less item
        return                             # exit immediately
        raise ValueError('value not found')    # only reached if no match
```

```python
import sys

def get_size(obj, seen=None):
    """Recursively finds size of objects"""
    size = sys.getsizeof(obj)
    if seen is None:
        seen = set()
    obj_id = id(obj)
    if obj_id in seen:
        return 0
    # Important mark as seen *before* entering recursion to gracefully handle
    # self-referential objects
    seen.add(obj_id)
    if isinstance(obj, dict):
        size += sum([get_size(v, seen) for v in obj.values()])
        size += sum([get_size(k, seen) for k in obj.keys()])
    elif hasattr(obj, '__dict__'):
        size += get_size(obj.__dict__, seen)
    elif hasattr(obj, '__iter__') and not isinstance(obj, (str, bytes, bytearray)):
        size += sum([get_size(i, seen) for i in obj])
    return size
```
